```python
# Step 1: Data Loading & Exploration

# 1. Import the pandas library
import pandas as pd

# 2. Load the COVID-19 data directly from the web
# (This avoids download issues and always gives you the latest data)
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(url)

# 3. Confirm that data was loaded successfully
print(" COVID-19 dataset loaded successfully!\n")

# 4. Display number of rows and columns in the dataset
print(f" Dataset Dimensions: {df.shape[0]:,} rows and {df.shape[1]} columns\n")

# 5. Preview the first 5 rows of the dataset
print(" First 5 Rows of the Dataset:")
display(df.head())
```

```python
# 6. Show all column names in a table format for better readability
print(" Column Names:")

# Turn the column names into a DataFrame for cleaner display
columns_df = pd.DataFrame(df.columns, columns=["Column Name"])
display(columns_df)
df.columns.tolist()
```

```python
# 7. View summary information about the dataset
print(" Dataset Info (Data Types, Non-Null Values):")
df.info()
```

```python
# 8. Check how many missing values are in each column
print("Missing Values Per Column (only showing columns with missing data):")

# Create a DataFrame of missing value counts
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]  # Only show columns with missing data
missing_df = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage (%)": (missing_values / len(df)) * 100
}).sort_values(by="Missing Values", ascending=False)

# Display the missing values as a table
display(missing_df)
```

```python
# 9. Show summary statistics for all numeric columns
print("Summary Statistics (Numerical Data):")
display(df.describe())
```

```python
# -----------------------------------------------
# Step 3: Data Cleaning
# -----------------------------------------------

# 1. Convert the 'date' column to datetime format
# This allows us to plot and filter data by time
df['date'] = pd.to_datetime(df['date'])

# 2. Filter for countries of interest: Kenya, USA, and India
# We exclude aggregate regions like "World" or "Africa"
countries = ['Kenya', 'United States', 'India']
df_filtered = df[df['location'].isin(countries)].copy()

# 3. Drop rows that do not have a date or critical data
# We'll remove rows with missing 'total_cases' or 'total_deaths'
df_filtered = df_filtered.dropna(subset=['total_cases', 'total_deaths'])

# 4. Fill or interpolate remaining missing values in important numeric columns
# This ensures smooth trend lines in plots without gaps
numeric_cols = ['new_cases', 'new_deaths', 'total_vaccinations']
df_filtered[numeric_cols] = df_filtered[numeric_cols].fillna(method='ffill')

# 5. Reset the index for clean data
df_filtered.reset_index(drop=True, inplace=True)

# Show cleaned dataset preview
print(" Cleaned Data Preview:")
display(df_filtered.head())
```

```python
# Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
### Total COVID-19 Cases Over Time
This line chart shows how total COVID-19 cases have increased over time in Kenya, India, and the United States. Notice the early surge in the US and India compared to Kenya.
```

```python
# Line chart for total cases over time in selected countries

plt.figure(figsize=(12,6))
sns.lineplot(data=df_filtered, x="date", y="total_cases", hue="location", linewidth=2.5)

plt.title("Total COVID-19 Cases Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

```python
# Line chart for total deaths over time

plt.figure(figsize=(12,6))
sns.lineplot(data=df_filtered, x="date", y="total_deaths", hue="location", linewidth=2.5)

plt.title(" Total COVID-19 Deaths Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

```python
# Line chart for new daily cases

plt.figure(figsize=(12,6))
sns.lineplot(data=df_filtered, x="date", y="new_cases", hue="location", linewidth=2)

plt.title(" New Daily COVID-19 Cases", fontsize=16)
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

```python
# Create a new column for death rate = total_deaths / total_cases
df_filtered["death_rate"] = df_filtered["total_deaths"] / df_filtered["total_cases"]

# Plot death rate over time
plt.figure(figsize=(12,6))
sns.lineplot(data=df_filtered, x="date", y="death_rate", hue="location", linewidth=2)

plt.title(" COVID-19 Death Rate Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Death Rate (Proportion)")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

```python
# Line chart showing total vaccinations over time

plt.figure(figsize=(12,6))
sns.lineplot(data=df_filtered, x='date', y='total_vaccinations', hue='location', linewidth=2.5)

plt.title(" Total COVID-19 Vaccinations Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# Filter relevant data
vax_data = df[df['location'].isin(['Kenya', 'India', 'United States'])].copy()

# Drop rows with missing total_vaccinations
vax_data = vax_data.dropna(subset=['total_vaccinations'])

# Plot cumulative vaccinations over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=vax_data, x='date', y='total_vaccinations', hue='location')
plt.title(' Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get the latest available vaccination data per country
countries = ['Kenya', 'India', 'United States']
latest_vax_data = []

for country in countries:
    country_data = df[df['location'] == country]
    country_data = country_data.dropna(subset=['people_vaccinated_per_hundred'])
    if not country_data.empty:
        latest_row = country_data.sort_values('date', ascending=False).iloc[0]
        latest_vax_data.append(latest_row)

# Create DataFrame
latest_vax_df = pd.DataFrame(latest_vax_data)

# Plot
plt.figure(figsize=(7, 5))
sns.barplot(data=latest_vax_df, x='location', y='people_vaccinated_per_hundred', palette='Greens')
plt.title('💉 % of Population Vaccinated (Most Recent Available)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('% Vaccinated')
plt.tight_layout()
plt.show()
```

```python
#  Bar chart comparing latest total vaccinations by country

# Get the latest date's data for each country
latest_vax = df_filtered.sort_values('date').groupby('location').tail(1)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=latest_vax, x='location', y='total_vaccinations', palette='viridis')

plt.title("Final Total Vaccinations by Country", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Total Vaccinations")
plt.tight_layout()
plt.show()
```

```python
# List of countries
countries = ['Kenya', 'India', 'United States']

# Loop to create pie chart per country
for country in countries:
    # Filter and drop rows with missing data
    country_df = df[df['location'] == country].dropna(subset=['people_vaccinated_per_hundred', 'population'])
    
    if not country_df.empty:
        # Get the latest available data
        latest_row = country_df.sort_values('date', ascending=False).iloc[0]
        
        vaccinated_pct = latest_row['people_vaccinated_per_hundred']
        unvaccinated_pct = 100 - vaccinated_pct
        
        # Pie chart
        plt.figure(figsize=(5, 5))
        plt.pie(
            [vaccinated_pct, unvaccinated_pct],
            labels=['Vaccinated', 'Unvaccinated'],
            autopct='%1.1f%%',
            startangle=90,
            colors=['#2ecc71', '#e74c3c']
        )
        plt.title(f' Vaccination Distribution in {country}')
        plt.axis('equal')
        plt.tight_layout()
        plt.show()
    else:
        print(f"No vaccination data available for {country}.")
```

```python
# Sum total cases across selected countries
global_cases = df[df['location'].isin(['Kenya', 'India', 'United States'])]
total_cases_by_country = global_cases.groupby('location')['total_cases'].max()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(total_cases_by_country, labels=total_cases_by_country.index,
        autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title(' Proportion of Total Cases by Country')
plt.tight_layout()
plt.show()
```

```python
# Filter the data for the specified countries
latest_stats = df[df['location'].isin(['Kenya', 'India', 'United States'])].copy()

# Drop rows with missing values for total_cases or total_deaths
latest_stats.dropna(subset=['total_cases', 'total_deaths'], inplace=True)

# Remove rows with zero total cases to avoid divide-by-zero
latest_stats = latest_stats[latest_stats['total_cases'] > 0]

# Calculate death rate
latest_stats['death_rate'] = latest_stats['total_deaths'] / latest_stats['total_cases']

# Plot the death rate
plt.figure(figsize=(7, 5))
sns.barplot(data=latest_stats, x='location', y='death_rate', palette='Reds_r')
plt.title(' COVID-19 Death Rate by Country (Latest Data)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Death Rate (Deaths per Case)')
plt.tight_layout()
plt.show()
```

```python
# Base filter
df_sel = df[df['location'].isin(['Kenya', 'India', 'United States'])].copy()
df_sel.dropna(subset=[
    'new_deaths_smoothed', 
    'new_vaccinations_smoothed', 
    'positive_rate',
    'total_tests_per_thousand',
    'total_cases_per_million'
], inplace=True)

# 1. New daily deaths (7-day average)
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sel, x='date', y='new_deaths_smoothed', hue='location')
plt.title('New Daily COVID-19 Deaths (7-day avg)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Deaths per Day (Smoothed)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. New daily vaccinations (7-day average)
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sel, x='date', y='new_vaccinations_smoothed', hue='location')
plt.title('New Daily COVID-19 Vaccinations (7-day avg)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Vaccinations per Day (Smoothed)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Positive test rate over time
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sel, x='date', y='positive_rate', hue='location')
plt.title('COVID-19 Test Positivity Rate Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Positive Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Total tests per thousand over time
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sel, x='date', y='total_tests_per_thousand', hue='location')
plt.title('Total COVID-19 Tests per Thousand Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Tests per Thousand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Total cases per million over time
plt.figure(figsize=(8,5))
sns.lineplot(data=df_sel, x='date', y='total_cases_per_million', hue='location')
plt.title('Total COVID-19 Cases per Million Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Cases per Million')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# a) Death Rate vs. Vaccinations (per hundred)
plt.figure(figsize=(7,5))
plt.scatter(
    df_sel['total_vaccinations_per_hundred'], 
    df_sel['death_rate'], 
    alpha=0.6
)
plt.title('Death Rate vs. Vaccinations per 100 People')
plt.xlabel('Total Vaccinations per Hundred')
plt.ylabel('Death Rate (Deaths per Case)')
plt.tight_layout()
plt.show()

# b) New Cases vs. New Deaths (both smoothed)
plt.figure(figsize=(7,5))
plt.scatter(
    df_sel['new_cases_smoothed'], 
    df_sel['new_deaths_smoothed'], 
    alpha=0.6
)
plt.title('New Daily Cases vs. New Daily Deaths (7-day avg)')
plt.xlabel('New Cases (Smoothed)')
plt.ylabel('New Deaths (Smoothed)')
plt.tight_layout()
plt.show()

# c) Tests per Thousand vs. Positive Rate
plt.figure(figsize=(7,5))
plt.scatter(
    df_sel['total_tests_per_thousand'], 
    df_sel['positive_rate'], 
    alpha=0.6
)
plt.title('Testing Intensity vs. Test Positivity Rate')
plt.xlabel('Total Tests per Thousand')
plt.ylabel('Positive Test Rate')
plt.tight_layout()
plt.show()
```

```python
# Heatmaps
```

```python
# Select only the numeric columns you care about
corr_cols = [
    'total_cases', 'total_deaths', 'new_cases_smoothed', 'new_deaths_smoothed',
    'death_rate', 'total_vaccinations', 'total_vaccinations_per_hundred',
    'total_tests_per_thousand', 'positive_rate'
]
corr_df = df_sel[corr_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(
    corr_df,
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)
plt.title('Correlation Matrix of COVID-19 Metrics', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Pivot to have dates on the x-axis and countries on the y-axis
pivot = df_sel.pivot_table(
    index='location',
    columns='date',
    values='new_cases_smoothed',
    aggfunc='mean'
)

plt.figure(figsize=(12,4))
sns.heatmap(
    pivot,
    cmap='Reds',
    cbar_kws={"label": "New Cases (7-day avg)"},
    xticklabels=20  # show every 20th date label
)
plt.title('Heatmap of New Daily Cases (Smoothed) Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
